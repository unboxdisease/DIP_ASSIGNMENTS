```python
import matplotlib.pyplot as plt
from matplotlib import colors
import cv2 
import numpy as np
```

```python
roads = cv2.imread("../images/roads.jpg",0)
plt.figure(figsize = (10,10)) 
plt.imshow(roads,cmap='gray')
plt.title("Non Overlappig Coins")
plt.show()
```

```python
def NormalThresholding(image,T):
    im = image.copy()
    for x in range(im.shape[0]):
        for y in range(im.shape[1]):
            if image[x][y] > T:
                im[x][y] = 255
            else:
                im[x][y] = 0
    return im

roads = cv2.imread("../images/roads.jpg",0)
roads = NormalThresholding(roads,125)
            
kernel = np.ones((2,2),np.uint8)
roads = cv2.dilate(roads,kernel,iterations = 3)

kernel = np.ones((4,4),np.uint8)
roads = cv2.erode(roads,kernel,iterations = 2)

kernel = np.ones((2,2),np.uint8)
roads = cv2.dilate(roads,kernel,iterations = 5)

roads = cv2.medianBlur(roads,11)
plt.imshow(roads,cmap='gray')
plt.show()
```

```python
edges = cv2.Canny(roads,100,200)
image = np.zeros(edges.shape)
image[:,550:690] = edges[:,550:690]
plt.imshow(image,cmap='gray')
plt.show()
```

```python
roads = cv2.imread("../images/roads.jpg",0)
roads = NormalThresholding(roads,125)
            
kernel = np.ones((2,2),np.uint8)
roads = cv2.dilate(roads,kernel,iterations = 3)

kernel = np.ones((4,4),np.uint8)
roads = cv2.erode(roads,kernel,iterations = 2)

kernel = np.ones((2,2),np.uint8)
roads = cv2.dilate(roads,kernel,iterations = 8)

roads = cv2.medianBlur(roads,7)
image = roads.copy()
image = 255- image
# output = roads.copy()
output = cv2.imread("../images/roads.jpg")
circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 100,
                          param1=50, param2=30, minRadius=200, maxRadius=500)
detected_circles = np.uint16(np.around(circles))
color = (255, 0, 0)

for (x, y ,r) in detected_circles[0, :]:
    cv2.circle(output, (x, y), r, (0, 255, 0), 7)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()
```
