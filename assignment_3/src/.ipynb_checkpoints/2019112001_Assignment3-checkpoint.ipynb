```python
def medianfilter1(image, k):
    if(k%2==0):
        t=k/2
    else:
        t=math.floor(k/2)
    image = np.pad(image,((t,t),(t,t)),'constant')
    out = np.zeros(image.shape)
    for i in range(t,image.shape[0]-t):
        for j in range(t,image.shape[1]-t):
            temp = image[i-t:i+t+1,j-t:j+t+1]
            tmp = temp.flatten()
            tmp.sort()
            out[i][j] = tmp[math.floor((k*k)/2)]
    return out

def medianFilter(img,k):
  r=img[:,:,0]
  g=img[:,:,1]
  b=img[:,:,2]

  img1 = medianfilter1(r, k)
  img2 = medianfilter1(g, k)
  img3 = medianfilter1(b, k)
  out = cv2.merge([img1, img2, img3])
  return out
```

```python
def colorLinContrastStretching(img, a, b):
  row = img.shape[0]
  col = img.shape[1]
  height = img.shape[2]
  maxi = np.max(img)
  mini = np.min(img)

  for i in range (row):
    for j in range (col):
        for k in range (height):
          img[i][j][k] = a + (((b-a)/(maxi-mini))*(img[i][j][k]-mini))
  
  return img
```

```python
def histEqualization1(image):
    p = {}
    hist=np.zeros(256)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            hist[image[i][j]]+=1
    a=np.zeros(hist.shape)
    a[0]=hist[0]
    for i in range(1,hist.shape[0]):
        a[i]=a[i-1]+hist[i]
    for i in range(0,256):
        p[i] = int((a[i]-a.min())*255/(a.max()-a.min()))
    a = (a-a.min())*255/(a.max()-a.min())
    out=a[image]
    return p,out

def histEqualization(img):
  r=img[:,:,0]
  g=img[:,:,1]
  b=img[:,:,2]

  _,r = histEqualization1(r)
  _,g = histEqualization1(g)
  _,b = histEqualization1(b)

  out = cv2.merge([r, g, b])
  return out
```

```python
def vintage(img):
  r = img[:, :, 0]
  g = img[:, :, 1]
  b = img[:, :, 2]

  imgR = min(0.393*r + 0.769*g + 0.189*b, 255)
  imgG = min(0.349*r + 0.686*g + 0.168*b, 255)
  imgB = min(0.272*r + 0.534*g + 0.131*b, 255)
  
  output = np.stack((imgR, imgG, imgB), axis = 2)

  return output.astype(np.uint8)
```

```python
def matrix(img):
  row = img.shape[0]
  col = img.shape[1]
  for i in range(row):
    for j in range(col):
      img[i][j][0] = 255*((img[i][j][0]/255)**np.float(1.5))
      img[i][j][1] = 255*((img[i][j][1]/255)**np.float(0.8))
      img[i][j][2] = 255*((img[i][j][2]/255)**np.float(1.5))

  return img
```

```python
def vignette(image):
  output = np.copy(image)
  row = image.shape[0]
  col = image.shape[1]
  height = image.shape[2]

  x_kernel = cv2.getGaussianKernel(col,200)
  y_kernel = cv2.getGaussianKernel(row,200)
  final = y_kernel*x_kernel.T

  mask = 255*final/np.linalg.norm(final)

  for i in range(height):
      output[:,:,i] = output[:,:,i] * mask

  return output
```

```python
def duotone(img, number):
    for i in range(3):
        if i == number:
            table = np.array([min((j**1.05), 255) for j in np.arange(0, 256)]).astype("uint8")
            img[:, :, i] = cv2.LUT(img[:,:,i], table)
        else:
            img[:, :, i] = 0 
    return img
```

```python
def union(a, parent):
    if parent[a] != a:
        parent[a] = union(parent[a], parent)
    return parent[a]

def ConnectedComponents(img):
    row = img.shape[0]
    col = img.shape[1]

    parent = [0]
    nextLabel = 1
    
    labels = np.zeros(img.shape, dtype=int)
    
    for h in range(row):
        for w in range(col)):
            if img[h, w] != 0:
                neighbors = []
                if h > 0:
                    if img[h-1][w] == img[h, w]:
                        neighbors.append((h-1, w))
                if h > 0 and w + 1 < col:
                    if img[h-1, w+1] == img[h, w]:
                        neighbors.append((h-1, w+1))
                if h > 0 and w > 0:
                    if img[h-1, w-1] == img[h, w]:
                        neighbors.append((h-1, w-1))
                if w > 0:
                    if (img[h, w-1] == img[h, w]):
                        neighbors.append((h, w-1))

                if len(neighbors) == 0:
                    parent.append(nextLabel)
                    labels[h, w] = nextLabel
                    nextLabel += 1                
                else:
                    L = [labels[r, c] for r, c in neighbors]
                    labels[h, w] = min(L)
                    for label in L:
                        parent[union(label, parent)] = union(labels[h, w], parent)
                        
    for h in range(row):
        for w in range(col):
            if img[h, w] != 0:
                labels[h, w] = union(labels[h, w], parent)
    
    return labels
```

```python
def dilating(img, k):
    row,col = img.shape
    img_copy = np.copy(img)
    for i in range(k//2,row-k//2):
        for j in range(k//2,col-k//2):
            w = img[i-k//2:i+k//2+1,j-k//2:j+k//2+1]
            img_copy[i][j] = np.max(w)
    img_copy = img_copy.astype(np.uint8)
    return img_copy

def Thresholding(img, threshold):
    output = np.copy(img)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j]> threshold:
                output[i][j] = 255
            else:
                output[i][j] = 0
    return output
```

```python
img = cv2.cvtColor(cv2.imread('coins.png'), cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()

threshold = [100,125,150]
for t in threshold:
    thresh = Thresholding(img,t)
    plt.subplot(1,2,1)
    plt.imshow(thresh, cmap='gray')
    plt.axis('off')
    plt.title('Threshold : ' + str(t))
    k = 9
    kernal = np.ones((k, k), np.uint8)
    dilation = dilating(thresh, k)
    plt.subplot(1,2,2)
    plt.imshow(dilation, cmap='gray')
    plt.title('Threshold : ' + str(t))
    plt.show()
    contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    objects = str(len(contours))
    text = "No. of Objects : "+str(objects)
    print(text)
```

```python
def otsu(img):
  hist, bin_centers_old = np.histogram(img.ravel(), 256) 
  hist = hist.astype(float) 
  bin_centers = np.zeros((bin_centers_old.shape[0] - 1))
  for i in range(len(bin_centers)-1):
      bin_centers[i] = (bin_centers_old[i]+bin_centers_old[i+1])/2
  weight1 = np.cumsum(hist) 
  weight2 = np.cumsum(hist[::-1])[::-1] 
  mean1 = np.cumsum(hist * bin_centers) / weight1 
  mean2 = (np.cumsum((hist * bin_centers)[::-1]) / weight2[::-1])[::-1] 
  variance12 = weight1[:-1] * weight2[1:] * (mean1[:-1] - mean2[1:]) ** 2 
  idx = np.argmax(variance12) 
  threshold = bin_centers[:-1][idx] 
  return threshold

val = otsu(img)
print('Otsu Threshold = ' + str(val))

thresh = Thresholding(img,val)
kernal = np.ones((9, 9), np.uint8)
dilation = dilating(thresh, 7)
plt.imshow(dilation, cmap='gray')
plt.axis('off')
plt.show()
contours, hierarchy = cv2.findContours(
    dilation, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

objects = str(len(contours))
print("No. of Objects:"+str(objects))
```

```python
def conv(image, k, kernel):
    if(k%2==0):
        t=math.floor(k/2)
        image = np.pad(image,((t-1,t-1),(t-1,t-1)),'constant')
    else:
        t=math.floor(k/2)
        image = np.pad(image,((t,t),(t,t)),'constant')
    out = np.zeros(image.shape)
#     print(image.shape)
    for i in range(t,image.shape[0]-t):
        for j in range(t,image.shape[1]-t):
            if(k%2==1):
                temp = image[i-t:i+t+1,j-t:j+t+1]    
            else:
                temp = image[i-(k-1)//2:i+t+1, j-(k-1)//2:j+t+1]
            out[i][j] = np.vdot(temp, kernel)
    return out
```

```python
def psf(img, sigma):
  row,col = img.shape
  center = (row/2,col/2)
  mask = np.zeros((row,col), dtype=np.float32)
  for i in range(row):
      for j in range(col):
          x1 = 2*np.pi*(sigma**2)
          x2 = np.exp(-(i**2 + j**2)/(2* sigma**2))
          mask[i][j]=(1/x1)*x2
  return mask
```

```python
def deblur(sigma):
    img1 = np.float32(img)/255
    
    psf_mask = psf(img1, sigma)
    
    img_fft1 = np.fft.fft2(img1)
    img_fft = np.fft.fftshift(img_fft1)

    psf_fft1 = np.fft.fft2(psf_mask)
    psf_fft = np.fft.fftshift(psf_fft1)

    res = img_fft/psf_fft

    out1 =  np.fft.ifftshift(res)
    out = np.fft.ifft2(out1).real
    
    return out
```

```python
from ast import literal_eval
with open('shinchan.pkl', 'rb') as f:
    my_string = pickle.load(f)

my_string = np.array(literal_eval(my_string))
plt.imshow(my_string)
plt.show()
a = my_string
my_string=str(my_string.tolist())
letters = []
only_letters = []

for letter in my_string:
    if letter not in letters:
        frequency = my_string.count(letter)             #frequency of each letter repetition
        letters.append(frequency)
        letters.append(letter)
        only_letters.append(letter)
        
print(letters)
```

### Part 1B

```python
nodes = []
while len(letters) > 0:
    nodes.append(letters[0:2])
    letters = letters[2:]                               # sorting according to frequency
nodes.sort()
print(nodes)
huffman_tree = []
huffman_tree.append(nodes) 

def combine_nodes(nodes):
    pos = 0
    newnode = []
    if len(nodes) > 1:
        nodes.sort()
        nodes[pos].append("1")                       # assigning values 1 and 0
        nodes[pos+1].append("0")
        combined_node1 = (nodes[pos] [0] + nodes[pos+1] [0])
        combined_node2 = (nodes[pos] [1] + nodes[pos+1] [1])  # combining the nodes to generate pathways
        newnode.append(combined_node1)
        newnode.append(combined_node2)
        newnodes=[]
        newnodes.append(newnode)
        newnodes = newnodes + nodes[2:]
        nodes = newnodes
        huffman_tree.append(nodes)
        combine_nodes(nodes)
    return huffman_tree                                     # huffman tree generation

newnodes = combine_nodes(nodes)
huffman_tree.sort(reverse = True)
print(huffman_tree)
checklist = []
for level in huffman_tree:
    for node in level:
        if node not in checklist:
            checklist.append(node)
        else:
            level.remove(node)
count = 0
for level in huffman_tree:
    print("Level", count,":",level)             #print huffman tree
    count+=1
print()
```
