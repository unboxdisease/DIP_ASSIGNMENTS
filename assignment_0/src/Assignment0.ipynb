```python
import cv2 
import matplotlib.pyplot as plt
img1 = cv2.imread("../images/lena.png")
img2 = cv2.imread("../images/original.png")
img3 = cv2.imread("../images/tower.jpg")
img4 = cv2.imread("../images/lotus.jpg")
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('2x2 subplot of images')
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img3)
ax4.imshow(img4)
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('2x2 subplot of images (Fixed)')
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img3)
ax4.imshow(img4, cmap = "gray")
```

```python
gray=0.2989*img3[:,:,0]+0.5870*img3[:,:,1]+0.1140*img3[:,:,2];
plt.imshow(gray,cmap='gray')
cv2.imwrite('../images/gray.jpg', gray)
```
