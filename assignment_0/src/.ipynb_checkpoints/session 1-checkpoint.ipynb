### Increase intensity of lena.png by 45 in grayscale

```python
img = cv2.imread("lena.png",cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape[0])
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    img[i,j]+=45

    
plt.imshow(img)
```

### Corrected method for linear intensity increase (What might cause an error on direct linear increase of intensities?)

```python
img = cv2.imread("lena.png",cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

print(img.shape[0])
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if img[i,j]+45 > 255:
      img[i,j] = 255
    else:  
      img[i,j]+=45

    
plt.imshow(img,cmap="gray")
```

### Construct a frequency histogram for bell.jpg, and plot the histogram using plt.bar (search on google for syntax)

```python
bell = cv2.imread("bell.jpeg",cv2.IMREAD_GRAYSCALE)
freq = []
for i in range(0,256):
  freq.append(0)
for i in range(bell.shape[0]):
  for j in range(bell.shape[1]):
    freq[int(bell[i,j])]+=1
plt.bar(range(0,256),freq)
```

### Perform thresholding on bell.jpg using the above constructed histogram. 

```python
for i in range(bell.shape[0]):
  for j in range(bell.shape[1]):
    if bell[i,j] > 127:
      bell[i,j] = 255
    else:
      bell[i,j] = 0
plt.imshow(bell,cmap="gray")


```

### Do a simple thresholding on the image sudoku.jpg, and find an apt threshold for the same using the above method. 

```python
sudoku = cv2.imread("sudoku.jpg",cv2.IMREAD_GRAYSCALE)
for i in range(sudoku.shape[0]):
  for j in range(sudoku.shape[1]):
    if sudoku[i,j] > 100:
      sudoku[i,j] = 255
    else:
      sudoku[i,j] = 0
plt.imshow(sudoku,cmap="gray")
```

### Challenge: Try chroma keying of images as a homework. Take fg.jpg foreground image and put it on bg.jpg. 

```python
fg = cv2.imread("fg.jpg",cv2.IMREAD_UNCHANGED)
image_copy = fg
lower_g = np.array([0, 150, 0])     ##[R value, G value, B value]
upper_g = np.array([101, 255, 101 ])
mask = cv2.inRange(image_copy, lower_g, upper_g)
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0, 0, 0]
background_image = cv2.imread('bg.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

crop_background = background_image[0:720, 0:1280]

crop_background[mask == 0] = [0, 0, 0]
final_image = crop_background + masked_image
plt.imshow(final_image)
```
