```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Affine Transform boy_smiling.jpg by height/4 in the x direction and height/5 in y direction. 

```python
boy = cv2.imread("boy_smiling.jpg",cv2.IMREAD_UNCHANGED)
boy = cv2.cvtColor(boy,cv2.COLOR_BGR2RGB)
h,w,c = boy.shape
hx = h/4
hy = h/5
T = np.float32([[1,0,hx],[0,1,hy]])
img = cv2.warpAffine(boy,T,(w+200,h+150))
plt.imshow(img)
```

### Read ariana_aligned.png from the images. Notice that the eyes are horizontal with respect to the screen. Your goal is to Affine Transform the given image such that the image is horizontal with respect to the screen. As shown in the image below. (Hint, manually find eye pixel coordinates and coordinates for the bottom corners of the image. Then use cv2.getTransform2D and cv2.warpAffine)
ariana.png

```python
ariana = cv2.imread("ariana_aligned.png")
ariana = cv2.cvtColor(ariana,cv2.COLOR_BGR2RGB)
h,w,c = ariana.shape
rot_mat = cv2.getRotationMatrix2D( (50,300), -45, 1 )
new = cv2.warpAffine(ariana,rot_mat,(h+100,w+100))
plt.imshow(new)

    
```

### Apply a homography on coca-cola.jpg and place it on the stadium.jpg such that the placement is perspective-appropriate. (Find corner coordinates for the coco-cola image, and the corners for placement in the stadium image. Then use cv2.findHomography and cv2.warpPerspective.)


```python
coke = cv2.imread("coca-cola.jpg")
coke = cv2.cvtColor(coke,cv2.COLOR_BGR2RGB)
coke.shape
pts1 = np.float32([[0 ,0],[820 ,0],[0, 260],[820, 260]])
pts2 = np.float32([[200,1000],[800,1000],[0,1200],[1000,1200]])
h, mask = cv2.findHomography(pts1, pts2, cv2.RANSAC,5.0)
stad = cv2.imread("stadium.jpg")
stad = cv2.cvtColor(stad,cv2.COLOR_BGR2RGB)

height, width ,cc = stad.shape
im1Reg = cv2.warpPerspective(coke, h, (width, height))
plt.imshow(im1Reg)
```
