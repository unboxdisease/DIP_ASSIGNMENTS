```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = []
img.append(cv2.imread('IIIT.webp'))
img.append(cv2.imread('IIIT2.jpeg'))
img.append(cv2.imread('IIIT3.jpeg'))
img.append(cv2.imread('IIIT4.webp'))
```

```python
for i in range(0,4):
    img[i] = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
```

```python
f,matrix=plt.subplots(2,2)
img_cnt = 0
for i in range(0,2):
    for j in range(0,2):
            matrix[i,j].imshow(img[img_cnt])
            img_cnt += 1
```

```python
gray=0.2989*img[0][:,:,0]+0.5870*img[0][:,:,1]+0.1140*img[0][:,:,2];
plt.imshow(gray,cmap='gray')
cv2.imwrite('images/gray_IIIT.jpg', gray)
```
